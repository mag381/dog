```python
import cv2
from matplotlib import pyplot as plt
!pip install icrawler
from icrawler.builtin import BingImageCrawler

from PIL import Image
import os, glob
import numpy as np
from PIL import ImageFile
# IOError: image file is truncated (0 bytes not processed)回避のため
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
#  画像取得始まり
key_image = "馬"    # 取得したい画像のキーワード
num_image = 6       # 取得する画像の数
dir_name = "horse"    # 画像を保存するフォルダ名
photos_dir = "./"+dir_name

# 猫の画像を100枚取得
crawler = BingImageCrawler(storage={"root_dir": dir_name})
crawler.crawl(keyword=key_image, max_num=num_image)
```

```python
# 取得画像の表示
num_data = 6  #　たくさん表示すると、一つ一つが小さく表示されるので 6 にしておく


files = glob.glob(photos_dir + "/*.jpg")

print(files)
for i, file in enumerate(files):        # files に該当する file を順番に取り出す
    image = Image.open(file)            # 画像ファイルを開く
    image = image.convert("RGB")        # カラーをRGB形式に変換
    # image = image.resize((image_size, image_size))
    ax = plt.subplot(num_data // 3, 3, i+1)   # 画像を行、列に配置する　3なら1行に3画像を配置する
    plt.tight_layout()                  # 画像同士が重ならないようにする
    file_name=os.path.split(file)[1]
    #file_name=os.path.basename(file)   # ファイル名のみの取得は　split でも　basename　でもよい
    ax.set_title(file_name)             # ファイル名表示の準備
    #ax.set_title(str(i+1))
    plt.imshow(image)                   # 画像を表示
```
