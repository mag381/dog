```python
# OpenCV  Canny   Monochrome version

import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab import files
uploaded_file = files.upload()
uploaded_file_name = next(iter(uploaded_file))
orig = cv2.imread(uploaded_file_name)
src = cv2.cvtColor(orig, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(orig, cv2.COLOR_BGR2GRAY)
plt.imshow(orig, cmap='gray',vmax=255,vmin=0)
plt.title('Original')
```

```python
edges = cv2.Canny(orig,100,200)   # エッジを決めるしきい値　100:黒から白への変化、200:白から黒への変化の境界値
plt.figure(figsize=(8,16))
plt.subplot(1,2,1)                # 1行に2列を表示するうちの1つ目を指定
plt.imshow(orig,cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)                # 1行に2列を表示するうちの2つ目を指定
plt.imshow(edges,cmap='gray')
plt.title('Canny edge')
```

```python
edges2 = cv2.Canny(orig,50,80)
plt.figure(figsize=(8,16))
plt.subplot(1,2,1)
plt.imshow(orig,cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(edges2,cmap='gray')
plt.title('Canny edge')
```
